```python
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0.1,1-1e-10, 100)

def mse(p, y):
    #y = np.where(p>0.5, 1, 0)
    return -(y-p)**2

def ce(p, y):
    #y = np.where(p>0.5, 1, 0)
    return y*np.log(p) + (1-y)*np.log(1-p)

l1 = mse(p, 1)
l2 = ce(p, 1) 

l1 = -l1/min(l1)
l2 = -l2/min(l2)

plt.plot(p, l1, label='MSE')
plt.plot(p, l2, label='cross-entropy')
plt.grid()
plt.xlabel('h(x)')
plt.ylabel('Error')
plt.legend()
    
```

```python
import scipy.stats

def gaussian_pdf(p, y):
    sd = np.sqrt(p*(1-p))
    return scipy.stats.norm.logpdf(y, loc=p, scale=sd)
    
llh1 = gaussian_pdf(p, 1)    
llh2 = ce(p,1)

plt.plot(p, llh1, label='Gaussian')
plt.plot(p, llh2, label='Bernoulli')
plt.grid()
plt.xlabel('True Probability')
plt.ylabel('Log-likelihood')
plt.legend()
    
```
